```python
''' 실패작1 '''
def dfs(a, idx, visit, suma, target):
    visit[idx] = True
    
    print(idx)
    print(visit, suma, len(visit))
    if (idx+1) == len(visit): return 1
    
    if not visit[idx+1]:
        suma += a[0]
        dfs(a, idx+1, visit, suma, target)
    else:
        if suma == target:        return 1
        else: return 0
    return 0

def solution(numbers, target):
    answer = 0
    
    arr = [[i, i*(-1)] for i in numbers]
    visit = [False] * len(numbers)
    
    for idx, a in enumerate(arr[:-1]):
        suma = arr[-1][0]
        if not visit[idx]:
            answer += dfs(a, idx, visit, suma, target)
        print(answer)
        suma = arr[-1][1]
        if not visit[idx]:
            answer += dfs(a, idx, visit, suma, target)
        print(answer)
    
    print(arr)
    
    
    return answer
```

```python
''' 블로그 솔루션 - DFS(Stack 사용) - [1,-1]중 -1부터 계산함 '''
def solution(numbers, target):
    answer = 0    
    n = len(numbers)
    
    # 0번째 인덱스의 값을 먼저 [4과 -4]로 만든 후 stack이니까 -4부터 계산 ㄱ
    que = [[numbers[0], 0], [-1 * numbers[0], 0]]  
    
    while que:
        print(que)
        temp, idx = que.pop()
        idx += 1
        
        if idx < n:
            que.append([temp + numbers[idx], idx])
            que.append([temp - numbers[idx], idx])
        else: # (idx == n 마지막까지 가면)
            if temp == target:
                answer += 1
                print('---------------')
    
    return answer
```

```python
solution([1, 1, 1, 1, 1], 3)
```

```python
solution([4, 1, 2, 1], 4)
```

```python
''' 블로그 답 - BFS - 그냥 계산한거 stack에 누적시켜서 count 사용'''
def solution(numbers, target):
    sup= [0]
    print(sup)
    for i in numbers:
        sub_node = []
        for j in sup : 
            sub_node.append(j+i)
            sub_node.append(j-i)
        sup = sub_node
        print(sub_node)
        
    return sup.count(target)
```

```python
solution([1, 1, 1, 1, 1], 3)
```

```python
solution([4, 1, 2, 1], 4)
```
