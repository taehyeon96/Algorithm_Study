```python
def solution(new_id):
    stopword = "~!@#$%^&*()=+[{]}:?,<>/"   
    
    # 1단계
    tmp = list(new_id.lower())
    
    # 2단계
    tmp2 = [data for data in tmp if data not in stopword]
    
    # 3단계
    answer = []
    tmp4 = 0
    for i in tmp2:
        if i == '.':  tmp4 += 1
        else:
            if tmp4 > 0:
                answer.append('.')
                answer.append(i)
            else:
                answer.append(i)
            tmp4 = 0
            
    # 4단계
    if answer and answer[0] == '.': del answer[0]
    if answer and answer[-1] == '.': del answer[-1]
        
    # 5단계
    if not answer:
        answer.append('a')
        
    # 6단계
    if len(answer) >= 16: answer = answer[:15]
    if answer[-1] == '.': del answer[-1]
        
    # 7단계
    while(len(answer)<=2):
        answer.append(answer[-1])
    
    return "".join(answer)
```

```python
''' 다른 사람 코드 - 정규식 사용 (좋아요 226) '''
#  https://ponyozzang.tistory.com/335

import re

def solution(new_id):
    st = new_id
    st = st.lower()
    st = re.sub('[^a-z0-9\-_.]', '', st)
    st = re.sub('\.+', '.', st)
    st = re.sub('^[.]|[.]$', '', st)
    st = 'a' if len(st) == 0 else st[:15]
    st = re.sub('^[.]|[.]$', '', st)
    st = st if len(st) > 2 else st + "".join([st[-1] for i in range(3-len(st))])
    return st
```

```python
''' 다른 사람 코드 - 나랑 비슷 비교해보기 (좋아요 55) '''
''' 최대한 문자열 그대로 활용하는 코딩 연습 '''

def solution(new_id):
    answer = ''
    # 1
    new_id = new_id.lower()
    
    # 2
    for c in new_id:
        if c.isalpha() or c.isdigit() or c in ['-', '_', '.']:
            answer += c
                    
    # 3 : 반복되는 요소는 문자열에서 그대로 가져와서 replace() 함수 사용 (중요)
    while '..' in answer:
        answer = answer.replace('..', '.')
        
        
    # 4 : 제거를 문자열 슬라이싱[:]으로 했음 (중요)
    if answer[0] == '.':
        answer = answer[1:] if len(answer) > 1 else '.'
    if answer[-1] == '.':
        answer = answer[:-1]
        
        
    # 5
    if answer == '':
        answer = 'a'
        
    # 6 : 마찬가지로 제거는 문자열 슬라이싱[:] 사용
    if len(answer) > 15:
        answer = answer[:15]
        if answer[-1] == '.':
            answer = answer[:-1]
            
    # 7
    while len(answer) < 3:
        answer += answer[-1]
        
    return answer
```
