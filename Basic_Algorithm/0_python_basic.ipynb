* 딕셔너리의 key 또는 value만 추출<중요>

```python

dic = {
    '사과' : 'apple',
    '바나나' : 'banana',
    '코코넛' : 'Coconut'
}

# 키 데이터만 담은 리스트 생성
key_list = dic.keys()

# 값 데이터만 담은 리스트 생성
val_list = dic.values()

print(key_list)
print(val_list)

# key에 알맞은 value값 꺼내기
print(dic['바나나'], "\n")

# 각 키에 따른 값을 하나씩 출력
for i in key_list:
    print(dic[i])
```

## 반복문

```python
# while을 사용하여 홀짝 나누기 EX

i = 1

while i <= 10:
    if i % 2 == 0:
        print(str(i), "짝수")
    else:                          #elif i % 2 == 1:
        print(f"{i} 홀수")
    i += 1

```

```python

# 위 코드를 for문과 continue를 사용하여 1부터 9까지 홀수의 합을 구해라

result = 0
for i in range(1, 10):
    if i % 2 == 0:
        continue
    else:
        result += i
        
print(result)
```

```python
# EX1) 아래와 같이 점수가 입력되었을 때 각 학생의 합격 여부를 판단해라
# 합격 커트라인은 80점

score = [90, 85, 77, 65, 97]

for i in range(len(score)):
    if score[i] >= 80:
        print(f"{i+1}번 학생은 합격입니다.")
        
##################################

result1 = [i+1 for i in range(len(score)) if score[i]>=80]
print(result1, "번 학생들은 합격입니다!")
```

```python
# EX1의 코드에서 부정행위를 한 2, 4번 학생을 제외하고 결과를 출력하라

score = [90, 85, 77, 65, 97]
chceating = [2, 4]

for i in range(len(score)):
    if i+1 in chceating:
        continue
    elif score[i] >= 80:
        print(f"{i+1}번 학생은 합격입니다.")
        
```

## 함수 
* global에 대한 이해 중요!

```python
def add(a, b):
    print(a+b)
    
add(b=2, a=3)
add(3,2)
```

```python
a = 7

def func(b):
    global a  # 함수 밖에 있는 global 변수를 사용
    print(a + b)
    
func(3)

#####################
array = [1,4,5,6,7]

def func(b):
    array.append(b)     # 함수 밖에 있는 list를 사용 가능
    print(array)
    
func(3)

#####################
array = [1,4,5,6,7]

def func(b):
    array = [1,2]
    array.append(b)     # 함수 밖에 있는 list를 사용하지 않음!!(우선순위!)
    print(array)
    
func(3)
```

## 람다 표현식

```python
def add(a,b):
    return a+b

print(add(3,7))

# lambda 매개변수: 반환값(매개변수)
print((lambda a, b: a+b)(3, 7))
```

```python
array = [('홍길동',50, 'A'), ('순신', 32, 'C'), ('무개', 74, 'B')]

def my_key(x):
    return x[2]

# 내장함수 sorted(정렬할리스트, key=정렬대상)
print(sorted(array, key=my_key))

print(sorted(array, key=lambda x: x[2]))
```

```python

# 람다 표현식을 여러 개의 리스트에 적용 <중요>

list1 = [1,2,3,4,5]
list2 = [6,7,8,9,9,10]

result = map(lambda a, b: a+b, list1, list2)

print(result)            # 이렇게 하면 출력이 안됨!
print(list(result))
```

## 순열과 조합



```python
""" 순열 (나열) """
from itertools import permutations

data = ['A', 'B', 'C']

print(list(permutations(data, 3)))
```

```python
""" 조합 """
from itertools import combinations

data = ['A', 'B', 'C']

print(list(combinations(data, 2)))
```

```python
""" Counter : 각 원소가 몇 번씩 등장했는지 출력 """

from collections import Counter

counter = Counter(['red', 'blue', 'red', 'green', 'blue', 'blue'])

print(counter['blue'])  # blue가 등장한 횟수 출력
print(counter['green'])

print(dict(counter))     # dictionary로 반환
```
